```python
student_performances = pd.read_csv('https://stepik.org/media/attachments/course/4852/StudentsPerformance.csv')
student_performances.head()
```

```python
student_performances = student_performances\
    .rename(columns={'parental level of education': 'parental_level_of_education',
                     'test preparation course': 'test_preparation_course',
                     'math score': 'math_score',
                     'reading score': 'reading_score',
                     'writing score': 'writing_score'})

student_performances.head()
```

```python
student_performances.query("writing_score > 75").head()
```

```python
student_performances.query("gender == 'female'").head()
```

```python
writing_score_query = 78
```

```python
student_performances.query(" (writing_score > @writing_score_query) & (gender == 'female') ").head()
```
