**Вопрос 1.**

Аналитик Ярослав захотел отфильтровать данные по комбинации условий и отобрать только те наблюдения, у которых значение по переменной writing score >  100, а значение переменной gender == female. Для этого Ярослав написал следующий запрос: 

students_performance[students_performance['writing score'] > 100 and students_performance.gender == 'female']


При работе с pandas series для логического И нужно использовать & вместо уже знакомого and

Нужно указать приоритет операций, взяв условия в скобки [(condition) и (condition)]

В python для объединения условий используются булевы операторы or и and. Сначала предполагается, что если используем энд, то справа и слева булевы выражения.

```python
True and False
```

Предположим, мы хотим скомбинировать 2 предыдущих примера. Как нам задать в этом случае общее условие фильтрации?

Сделать вот так не получится:

```python
students_performance.gender == 'female' and students_performance['writing score'] > mean_writing_score
```

Поскольку слева и справа от оператора and у нас не булевы примитивы (False или True), а серии pandas.

С ними нужно работать по-другому.

Преобразуем этот запрос:

каждое условие возьмём в скобки - изолируем, чтобы не было проблем 
and заменим на &
А результат запроса свяжем с переменной query.

```python
query = (students_performance['writing score'] > mean_writing_score) & (students_performance.gender == 'female')
```

```python
students_performance.loc[query].head()
```

Доп. вопрос: запрос с квадратными скобками, а не круглыми:

[students_performance.gender == 'female'] and [students_performance['writing score'] > mean_writing_score]
даёт вектор из True и False, но этот вектор отличается от вектора, получаемого при запросе:

(students_performance.gender == 'female') & (students_performance['writing score'] > mean_writing_score)
Почему так происходит?

Ответ:

Это:

(students_performance.gender == 'female') & (students_performance['writing score'] > mean_writing_score)
пандасовская векторизированная операция - все действия (сравнения, логическое и) выполняются поэлементно.

А с квадратными скобками вместо круглых, получается встроенный тип данных с python - список

[students_performance.gender == 'female'] # this is a list with pd.Series
Оператор and в питоне работает следующим образом - если 1-ый операнд True, то он возвращает 2-ой, в противном случае возвращается первый.

```python
[students_performance.gender == 'female'] and [students_performance['writing score'] > mean_writing_score]
```

```python
(students_performance.gender == 'female') & (students_performance['writing score'] > mean_writing_score)
```

```python
3 and 5
```

```python
0 and 3
```

У списков булевы значения такие:

[] => False
[что-нибудь] => True
Таким образом, мы просто получаем 2-ой список

Есть специальный метод query, также позволяющий нам фильтровать dataframe.

Но сначала давайте переименуем в нашем dataframe все колонки, содержащие символ пробела в своём имени.

Для этого выведем названия всех столбцов нашего dataframe и выберем названия с пробелами.

```python
#students_performance = pd.read_csv('https://stepik.org/media/attachments/course/4852/StudentsPerformance.csv')
students_performance = pd.read_csv('https://raw.githubusercontent.com/nedokormysh/Stepik_Karpov_ML/main/StudentsPerformance.csv')
students_performance.head()
```

```python
students_performance = students_performance\
    .rename(columns={'parental level of education': 'parental_level_of_education',
                     'test preparation course': 'test_preparation_course',
                     'math score': 'math_score',
                     'reading score': 'reading_score',
                     'writing score': 'writing_score'})

students_performance.head()
```

Метод query принимает строковый аргумент - запрос, по которому мы хотим получить нужные данные.

```python
students_performance.query("writing_score > 75").head()
```

Внутри нашего query можно без проблем проводить сравнения со строкой

```python
students_performance.query("gender == 'female'").head()
```

В query мы можем комбинировать наши запросы фильтрации

```python
students_performance.query('gender == "female" & math_score > 75').head()
```

Внутри query можно проводить сравнения с переменными, используя экранирование при помощи @

```python
writing_score_query = 78
```

```python
students_performance.query(" (writing_score > @writing_score_query) & (gender == 'female') ").head()
```
