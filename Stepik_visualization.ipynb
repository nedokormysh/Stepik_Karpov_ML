```python
#students_performance = pd.read_csv('https://stepik.org/media/attachments/course/4852/StudentsPerformance.csv')
students_performance = pd.read_csv('https://raw.githubusercontent.com/nedokormysh/Stepik_Karpov_ML/main/StudentsPerformance.csv')
students_performance.columns = [x.replace(" ", "_") for x in students_performance.columns]
```

```python
students_performance.math_score.mean()
```

```python
students_performance.math_score.mode()
```

Самый простой способ построить гистограммку, не выходя из датафрейма. Гистограмма распределения скоров по математике.

```python
students_performance.math_score.hist()
```

Пандас всё таки не библиотека для визуализации. И функция хист - это функция из матплотлиб

Нарисуем теперь скаттер-плот. График корреляции между двумя переменными.

```python
students_performance.plot.scatter(x='math_score', y='reading_score')
```

Для кастомизации лучше использовать библиотеку сиборн. Нарисуем скаттерплот при помощи сиборн. Для этого лмплот и есть несколько основных аргументов. икс, игрек - переменные, которые хотим отрисовать и данные, которые хотим использовать.

```python
sns.lmplot(x='math_score', y='reading_score', data=students_performance)
```

Здесь регрессионная прямая, которая показывает линейную аппроксимацию взаимосвязи этих двух переменных.

Можем сразу добавить на наш график группирующую переменную.

```python
sns.lmplot(x='math_score', y='reading_score', hue='gender', data=students_performance)
```

Покрасили наши точки в зависимости от переменной гендер. Добавили легенду на график и внутри каждой из групп добавили регрессионную прямую.

Если регрессионые пряммые не нужны.

```python
sns.lmplot(x='math_score', y='reading_score', hue='gender', data=students_performance, fit_reg=False)
```

Сразу наш график можно как-то кастомизировать. Добавить ему читаемости. Названия осей просто дублируют названия наших переменных. Например, нижнее подчёркивание в названиях осей не особо нужно. Сохраним наш график в переменную. У этого графика есть различные переменные, которые можно использовать.

```python
ax = sns.lmplot(x='math_score', y='reading_score', hue='gender', data=students_performance, fit_reg=False)
ax.set_xlabels('Math score')
ax.set_ylabels('Reading score')
```
